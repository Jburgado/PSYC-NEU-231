```python
#importing the seaborn package
import seaborn as sns
import matplotlib.pyplot as plt #maybe not necessary
```

```python
#loading tips dataset and visualizing data
tips = sns.load_dataset('tips')
tips.head()

#adding new column with percentage tip
tips['tip_percentage'] = 100 * tips['tip'] / tips['total_bill']

#visualize relationship between percentage with day of week and time
sns.pointplot(x='day',y='tip_percentage',hue='time',data=tips)
```

```python
#visualizing 3 variables - party size, day, time
sns.catplot(x='day',y='tip_percentage',hue='time',col='size',data=tips,kind='point')
plt.show()
```

```python
g = sns.PairGrid(tips, y_vars="tip_percentage",
                 x_vars=["size", "day", "time"],
                 height=5, aspect=.5)

# Draw a seaborn pointplot onto each Axes
g.map(sns.pointplot, scale=1.3, errwidth=4, color="xkcd:plum")
```
